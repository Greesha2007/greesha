```python
import pandas as pd

# Function to convert CGPA to GPA
def cgpa_to_gpa(cgpa):
    return (cgpa * 4.0) / 10.0

# Load the dataset
df = pd.read_csv('Data analyst.csv')  # Replace 'your_dataset.csv' with your actual file path

# Convert CGPA to GPA and create a new column 'GPA'
df['GPA'] = df['CGPA'].apply(cgpa_to_gpa)

# Save the updated dataset to a new CSV file (optional)
df.to_csv('updated_Data analyst.csv', index=False)

# Display the updated dataframe
print(df)
```

```python
import pandas as pd
Q1 = df['Designation'].quantile(0.25)
Q3 = df['Designation'].quantile(0.75)
IQR = Q3 - Q1

# Define outliers
outliers = (df['Designation'] < (Q1 - 1.5 * IQR)) | (df['Designation'] > (Q3 + 1.5 * IQR))

# Replace outliers with median
df.loc[outliers, 'Designation'] = df['Designation'].median()
```
