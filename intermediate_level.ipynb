```python
# Check the data type of the 'Diet_Type' column
print(df1['Diet_Type'].dtype)
```

```python
# Convert 'Diet_Type' column to string (if needed)
df1['Diet_Type'] = df1['Diet_Type'].astype(str)
```

```python
# Remove 'Diet:' from the 'Diet_Type' column
df1['Diet_Type'] = df1['Diet_Type'].str.replace('Diet:', '').str.strip()
```

```python
df1
```

```python
# Remove 'Course_name:' from the 'Course_name' column
df1['Course_name'] = df1['Course_name'].str.replace('Course:', '').str.strip()
```

```python
df1
```

```python
# Define categories based on the 'Course_name' column
def categorize_course(course_name):
    course = course_name.lower()
    if 'breakfast' in course:
        return 'Breakfasts'
    elif any(keyword in course for keyword in ['main course', 'lunch', 'one pot dish', 'dinner']):
        return 'Main Courses'
    elif any(keyword in course for keyword in ['appetizer', 'snack']):
        return 'Appetizers'
    elif 'dessert' in course:
        return 'Desserts'
    else:
        return 'Other'

# Apply the categorization function
df1['Category'] = df1['Course_name'].apply(categorize_course)

# Create the subset dataset
subset_data = df1[['name_of_Dish', 'Diet_Type', 'Course_name', 'Category', 'Discrption_of_Dish', 'Cuisine_name', 'Ratings_of_Dish', 'Similar_Dishes', 'Ingredients_of_Dish', 'Prepration_time', 'Cooking_time', 'Total_time', 'Makes', 'Recipe_Instructions']]

# Display the first few rows of the subset dataset
subset_data.head()
```

```python
# Convert the 'Ratings_of_Dish' column to string
df1['Ratings_of_Dish'] = df1['Ratings_of_Dish'].astype(str)

# Remove any extraneous characters and whitespace, then convert to float
df1['Ratings_of_Dish'] = df1['Ratings_of_Dish'].str.extract(r'(\d+\.\d+|\d+)').astype(float)

# Verify the conversion
print(df1.dtypes)
print(df1['Ratings_of_Dish'].head())
```

```python
# Fill missing values with an empty string for 'Similar_Dishes'
df1['Similar_Dishes'] = df1['Similar_Dishes'].fillna('')

# Remove square brackets and extra quotes from 'Similar_Dishes'
df1['Similar_Dishes'] = df1['Similar_Dishes'].str.replace(r"[\[\]']", "", regex=True)

# Remove square brackets and extra spaces from 'Cuisine_name'
df1['Cuisine_name'] = df1['Cuisine_name'].str.replace(r"[\[\]']", "", regex=True).str.strip()
```

```python
df1
```

```python
# Remove 'Cuisine:' from the 'Cuisine_name' column
df1['Cuisine_name'] = df1['Cuisine_name'].str.replace('Cuisine:', '').str.strip()
```

```python
df1
```

```python
def convert_time(time_str):
    # Remove 'M' and split the time string
    time_str = time_str.replace('M', '').strip()
    if 'H' in time_str:
        hours, minutes = time_str.split('H')
        minutes = int(hours) * 60 + int(minutes) if minutes else int(hours) * 60
    else:
        minutes = int(time_str)
    return minutes

# Apply the conversion function
df1['Prepration_time'] = df1['Prepration_time'].apply(convert_time)
df1['Cooking_time'] = df1['Cooking_time'].apply(convert_time)
df1['Total_time'] = df1['Total_time'].apply(convert_time)
```

```python
df1
```

```python
# Remove 'Servings' from the 'Makes' column
df1['Makes'] = df1['Makes'].str.replace('Servings', '').str.strip()
```

# Q1. How does the rating distribution differ between High protein vegetarian and high-protein non-vegetarian dishes?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create subsets for High Protein Vegetarian and High Protein Non Vegetarian
high_protein_vegetarian = df1[df1['Diet_Type'] == 'High Protein Vegetarian']
high_protein_non_vegetarian = df1[df1['Diet_Type'] == 'High Protein Non Vegetarian']

# Plot histograms
plt.figure(figsize=(12, 6))

# Plot for High Protein Vegetarian
plt.subplot(1, 2, 1)
plt.hist(high_protein_vegetarian['Ratings_of_Dish'], bins=10, edgecolor='black')
plt.title('High Protein Vegetarian Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')

# Plot for High Protein Non Vegetarian
plt.subplot(1, 2, 2)
plt.hist(high_protein_non_vegetarian['Ratings_of_Dish'], bins=10, edgecolor='black')
plt.title('High Protein Non Vegetarian Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# Q2. How does dish complexity vary across Highly Rated (>4.5) Indian and International Cuisines?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter highly rated dishes
highly_rated_df = df1[df1['Ratings_of_Dish'] > 4.5]

# Define complexity score
def calculate_complexity(row):
    num_ingredients = len(eval(row['Ingredients_of_Dish']))
    total_time = row['Prepration_time'] + row['Cooking_time']
    num_instructions = len(eval(row['Recipe_Instructions']))
    # Example complexity score (could be adjusted)
    complexity_score = num_ingredients + total_time / 10 + num_instructions
    return complexity_score

highly_rated_df['Complexity_Score'] = highly_rated_df.apply(calculate_complexity, axis=1)

# Create subsets for Indian and International cuisines
indian_cuisine_df = highly_rated_df[highly_rated_df['Cuisine_name'].str.contains('Indian', case=False)]
international_cuisine_df = highly_rated_df[~highly_rated_df['Cuisine_name'].str.contains('Indian', case=False)]

# Plot complexity scores
plt.figure(figsize=(12, 6))

# Plot for Indian Cuisine
plt.subplot(1, 2, 1)
plt.hist(indian_cuisine_df['Complexity_Score'], bins=10, edgecolor='black')
plt.title('Complexity of Highly Rated Indian Dishes')
plt.xlabel('Complexity Score')
plt.ylabel('Frequency')

# Plot for International Cuisine
plt.subplot(1, 2, 2)
plt.hist(international_cuisine_df['Complexity_Score'], bins=10, edgecolor='black')
plt.title('Complexity of Highly Rated International Dishes')
plt.xlabel('Complexity Score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Print out mean complexity scores for both categories
print(f"Mean Complexity Score for Indian Cuisine: {indian_cuisine_df['Complexity_Score'].mean()}")
print(f"Mean Complexity Score for International Cuisine: {international_cuisine_df['Complexity_Score'].mean()}")
```

# Q3. Calculate the Variation of cooking Time across various Vegetarian Diet_Types.

```python
import pandas as pd
import matplotlib.pyplot as plt


# Filter vegetarian dishes
vegetarian_df = df1[df1['Diet_Type'].str.contains('Vegetarian', case=False)]

# Calculate average cooking time for each vegetarian diet type
average_cooking_time = vegetarian_df.groupby('Diet_Type')['Cooking_time'].mean().reset_index()

# Plot the average cooking times
plt.figure(figsize=(10, 6))
plt.bar(average_cooking_time['Diet_Type'], average_cooking_time['Cooking_time'], color='skyblue')
plt.xlabel('Vegetarian Diet Type')
plt.ylabel('Average Cooking Time (minutes)')
plt.title('Average Cooking Time Across Various Vegetarian Diet Types')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Print out the average cooking times
print(average_cooking_time)
```

# Q4. How many dishes are Diabetic friendly and sugar-free?

```python
# Define keywords or patterns for diabetic-friendly and sugar-free dishes
diabetic_friendly_keywords = ['Diabetic Friendly']
sugar_free_keywords = ['Sugar Free']

# Filter dishes that match diabetic-friendly criteria
diabetic_friendly_df = df1[df1['Diet_Type'].str.contains('|'.join(diabetic_friendly_keywords), case=False, na=False)]

# Further filter for sugar-free dishes
diabetic_sugar_free_df = diabetic_friendly_df[diabetic_friendly_df['Diet_Type'].str.contains('|'.join(sugar_free_keywords), case=False, na=False)]

# Count the number of such dishes
num_diabetic_sugar_free_dishes = diabetic_sugar_free_df.shape[0]

# Print the number of dishes
print(f"Number of diabetic-friendly and sugar-free dishes: {num_diabetic_sugar_free_dishes}")
```

# Q5. What is the most common cuisine type for each course?

```python
# Group by 'Course_name' and 'Cuisine_name', and count occurrences
cuisine_counts = df1.groupby(['Course_name', 'Cuisine_name']).size().reset_index(name='Count')

# Find the most common cuisine for each course
most_common_cuisine = cuisine_counts.loc[cuisine_counts.groupby('Course_name')['Count'].idxmax()]

# Print the results
for _, row in most_common_cuisine.iterrows():
    print(f"The most common cuisine for {row['Course_name']} is {row['Cuisine_name']} with {row['Count']} dishes.")
```
